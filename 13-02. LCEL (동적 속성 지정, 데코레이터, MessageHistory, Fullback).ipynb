### 동적 속성 지정

```python
from dotenv import load_dotenv

load_dotenv()
```

- `configurable_fields` 메서드를 사용하여 `model_name` 속성을 동적 구성 가능한 필드로 지정

```python
from langchain_classic.prompts import PromptTemplate
from langchain_core.runnables import ConfigurableField
from langchain_openai import ChatOpenAI

model = ChatOpenAI(temperature=0, model_name="gpt-4o")

model.invoke("대한민국의 수도는 어디야?").__dict__
```

<br>

- `model.invoke()` 호출시 `config={"configurable": {"키": "값"}}` 형식으로 동적 지정

```python
model.invoke(
    "대한민국의 수도는 어디야?",
    config={"configurable": {"gpt_version": "gpt-3.5-turbo"}},
).__dict__
```

<br>

- `model` 객체의 `with_config()` 메서드를 사용하여 `configurable` 매개변수를 설정

```python
model.with_config(configurable={"gpt_version": "gpt-4o-mini"}).invoke(
    "대한민국의 수도는 어디야?"
).__dict__
```

<br>

- 함수를 체인의 일부로 사용할 때에도 동일한 방식으로 활용

```python
prompt = PromptTemplate.from_template("{x} 보다 큰 위의 난수를 생성")
chain = (
    prompt | model
)
```

```python
chain.with_config(configurable={"gpt_version": "gpt-4o"}).invoke({"x": 0}).__dict__
```

### `HubRunnable`: LangChain Hub의 설정 변경
- `HubRunnable` 을 사용하면 Hub 에 등록된 프롬프트의 전환을 용이


```python
from langchain_classic.runnables.hub import HubRunnable
```

```python
prompt = HubRunnable("teddynote/rag-prompt-korean").configurable_fields(
    # 소유자 저장소 커밋을 설정하는 ConfigurableField
    
    owner_repo_commit=ConfigurableField(
    
        # 필드의 ID
        id="hub_commit",
    
        # 필드의 이름
        name="Hub Commit",
    
        # 필드에 대한 설명
        description="Korean RAG prompt by teddynote",
    )
)
```

```python
prompt
```

<br>

- `with_config` 지정 없이 `prompt.invoke()` 메서드를 호출하면 처음 설정한 `"rlm/rag-prompt"` hub 에 등록된 프롬프트를 `pull` 하여 가져옴

```python
prompt.invoke({"question": "Hello", "context": "World"}).messages
```

```python
prompt.with_config(
    configurable={"hub_commit": "teddynote/simple-summary-korean"}
).invoke({"context": "Hello"})
```
